```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('startup_funding.csv')
```

```python
df.head(2)
```

```python
df.columns = df.columns.str.strip().str.replace(r'\s+', ' ', regex=True)
```

```python
df.drop(columns=['Remarks'], inplace=True)
```

```python
df['Investors Name'] = df['Investors Name'].fillna('Undisclosed')
df['SubVertical'] = df['SubVertical'].fillna('Unknown')
df['InvestmentnType'] = df['InvestmentnType'].fillna(df['InvestmentnType'].mode()[0])
```

```python
df['Amount in USD'] = (
    df['Amount in USD']
    .astype(str)
    .str.replace(',', '')
)
```

```python
df['Amount in USD'] = pd.to_numeric(df['Amount in USD'], errors='coerce')
```

```python
df = df.dropna(subset=['Amount in USD'])
```

```python
df['Amount_in_Cr'] = df['Amount in USD'] * 84 / 10000000
```

```python
df.set_index('Sr No', inplace=True)
```

```python
df.rename(columns={
    'Date dd/mm/yyyy': 'date',
    'Startup Name': 'startup',
    'Industry Vertical': 'vertical',
    'SubVertical': 'subvertical',
    'City Location': 'city',
    'Investors Name': 'investor',
    'InvestmentnType': 'round',
    'Amount in USD': 'amount'
}, inplace=True)
```

```python
df.head()
```

```python
df = df.drop(columns=['amount'])
```

```python
df['date'] = pd.to_datetime(df['date'], errors='coerce', dayfirst=True)
```

```python
df.info()
```

```python
df.to_csv('startup_cleaned.csv',index=False)
```
